```python
import pandas as pd
```

```python
clf_outpath = '../project/transactivity_error_analysis.csv'
clf_out = pd.read_csv(clf_outpath)
clf_out
```

```python
false_pos = clf_out[(clf_out['Predicted']=='y') & (clf_out['Gold']=='n')]
print(len(false_pos))

false_neg = clf_out[(clf_out['Predicted']=='n') & (clf_out['Gold']=='y')]
print(len(false_neg))
```

```python
cats = {}

for s in ['Predicted', 'Gold']:
    for b in ['y', 'n']:
        cats['{0}_{1}'.format(s.lower()[:4], b)] = clf_out[clf_out[s]==b]
        
cats.keys()
```

```python
avg_lens = {}

for cat in cats:
    avg_lens[cat] = len(' '.join(cats[cat]['Text'].values).split())/len(cats[cat])
avg_lens
```
