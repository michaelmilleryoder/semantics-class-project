```python
import os, csv
import pandas as pd
import numpy as np
```

```python
pdtb_dir = '/home/michael/school/11-727/project/pdtb_v2/pipes/'
```

```python
dms = set()
for i in range(25):
    i_format = '{:02d}'.format(i)
    print(i_format)
    for fname in os.listdir(os.path.join(pdtb_dir, i_format)):
        print(fname)
        if not fname.startswith('.'):
            df = pd.read_csv(os.path.join(pdtb_dir, i_format, fname), sep='|', header=None, quoting=csv.QUOTE_NONE,
                            encoding='utf8')
            file_dms = set(dm.lower() for dm in df[8].unique().tolist() if isinstance(dm, str))
            dms |= file_dms
    
dms
```

```python
with open('../project/pdtb_discourse_markers.txt', 'w') as f:
    for dm in sorted(dms):
        f.write(dm + '\n')
```
